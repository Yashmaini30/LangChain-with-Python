### Load Env Variables

```python
from dotenv import load_dotenv
load_dotenv() 
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite", temperature=0)
```

```python
prompts = [
    {"role":"user","content":"Who is yash maini"},
    {"role":"assistant","content":"Yash Maini is a data scientist and AI enthusiast known for his work in machine learning and artificial intelligence."},
    {"role":"user","content":"What is his favorite programming language?"}
]

res = llm.invoke(prompts)
print(res.content)
```

### Manage History

```python
history = []
while True:
    query = input("User: ")

    if query.lower() in ["exit", "quit", "bye"]:
        print("Exiting the chat. Goodbye!")
        break
    history.append({"role": "user", "content": query})
    print("User:", query)

    response = llm.invoke(history)

    history.append({"role": "assistant", "content": response.content})
    print("Assistant:", response.content, "\n")
```
