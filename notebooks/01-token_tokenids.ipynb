```python
text = "What a beautiful day!, she exclaimed."
```

```python
import tiktoken
```

```python
tokenizer = tiktoken.encoding_for_model("gpt-3.5-turbo")
```

```python
tokenIds = tokenizer.encode(text)
tokenIds
```

```python
resultIds = [2465, 836, 374, 379, 1003, 11, 323, 358, 1093, 311, 7172, 11033, 449, 1063, 5066, 13465]
result = tokenizer.decode(resultIds)
result
```
