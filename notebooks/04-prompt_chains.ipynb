### Prompt templates - Dynamic Prompting

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser    
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite", temperature=0)
out = StrOutputParser()
```

```python
### my self chain node

def to_upper_chain(input: str):
    return input.upper()
```

```python
prompts = ChatPromptTemplate.from_messages( [
    {"role":"system","content":"You are a {language} translator."},
    {"role":"user", "content":"{query}" }
])
```

```python
final_prompt = prompts.format_prompt(language="french", 
                                     query="i love yoou" )

res=llm.invoke(final_prompt)    
print(res.content)
```

### Chains--Runnable sequences of actions

```python
res = llm.invoke(prompts.invoke({ "language":"french",
                 "query":"i love yoou"})
)

res.content
```

```python
chains = prompts | llm | out | to_upper_chain  ## Chains--Runnable sequences of actions
```

```python
res = chains.invoke({ "language":"french",
                 "query":"i love yoou"})

print(res)
```
